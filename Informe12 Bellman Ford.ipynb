# **1. Descripción del Problema**

El problema a resolver en el informe consiste en encontrar el cámino máscorto donde se buscar recorrer la menor distancia a recorrer desde un nodo a otro.

El algoritmo que se presenta para resolver el problema es el de "Bellman Ford" el cual contiene los siguientes datos de entrada y salida:

$Entrada$: Un grafo dirigido $G=(V,E)$, un vértice fuente $s ϵ V$, y un valor real $l_e$ asociado a cada arco $e \in E$.

$Salida$: La distancia más corta $dist(s,v)$ para cada vértice $v \in V$.
Una declaración indicando que $G$ contiene un ciclo negativo.

# **2. Descripción del Algoritmo**

### **2.1 Bellman Ford**

El algoritmo de Bellman Ford permite encontrar el camino más corto entre nodos, incluyendo a su vez, la posibilidad de encontrar arcos **negativos** en la estructura del grafo .

El algoritmo desarrolla los siguientes pasos para su ejecución:


1. Inicializa un arreglo de distancias en infinito, exceptuando la del 
nodo inicial que es 0.

2. Actualiza las etiquetas $n-1$ veces, donde $n$ es la cantidad de nodos.

3. Comprueba la existencia de ciclos negativos.

4. Entrega el camino más corto entre la posición  inicial y todos los nodos, o un mensaje indicando la presencia de un ciclo negativo. 


### **2.1.1 Subestructura Óptima**

Se tienen 2 posibles casos:

1. La ruta más corta tiene $i-1$ arcos o menos, por lo que basta con encontrar la ruta más corta de $i-1$ arcos o menos.

2. La ruta tiene $i$ arcos por lo que se realiza subdivisión de subproblemas.

### **2.2 Algoritmo de Dijkstra**

El algoritmo de Dijkstra permite encontrar el camino más corto entre nodos, pero a diferencia de Bellman Ford, este requiere que todos sus arcos sean positivos aunque su tiempo de ejecución sigue siendo mejor que el del algoritmo mencionado anteriormente.

El algoritmo desarrolla los siguientes pasos para su ejecución:

Teniendo un grafo dirigido ponderado de  $N$ nodos no aislados, sea x el nodo inicial. Un vector $D$ de tamaño $N$ guardará al final del algoritmo las distancias desde $x$ hasta el resto de los nodos.

1. Inicializar todas las distancias en $D$ con un valor infinito relativo, ya que son desconocidas al principio, exceptuando la de $x$, que se debe colocar en $0$, debido a que la distancia de $x$ a $x$ sería $0$.

2. Sea $a=x$ Se toma $a$ como nodo actual.

3. Se recorren todos los nodos adyacentes de a, excepto los nodos marcados. Se les llamará nodos no marcados v$_i$.

4. Para el nodo actual, se calcula la distancia tentativa desde dicho nodo hasta sus vecinos con la siguiente fórmula: $dt(v_i) = Da + d(a,v_i)$. Es decir, la distancia tentativa del nodo $v_i$ es la distancia que actualmente tiene el nodo en el vector $D$ más la distancia desde dicho nodo $a$ (el actual) hasta el nodo $v_i$. Si la distancia tentativa es menor que la distancia almacenada en el vector, entonces se actualiza el vector con esta distancia tentativa. Es decir, si $dt(v_i) < D v_i → Dvi = dt(v_i)$

5. Se marca como completo el nodo $a$.

6. Se toma como próximo nodo actual el de menor valor en D (puede hacerse almacenando los valores en una cola de prioridad) y se regresa al paso 3, mientras existan nodos no marcados.

Una vez terminado al algoritmo, $D$ estará completamente lleno.

```python
import random as rd
import matplotlib.pyplot as plt
import datetime
import math as mt
from timeit import repeat
from termcolor import colored
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout
from sys import maxsize

def graph_to_nxdigraph(graph: list, n: int):
    """
        Input: Un grafo en formato list[tuple]. Ej: [(0, 1, 10), (1, 2, 15), (2, 0, 7)].
        Output: Un nx.DiGraph de la libreria networkx.
    """
    nxdigraph = nx.DiGraph()
    [nxdigraph.add_node(i) for i in range(n)]

    for v in graph:
        nxdigraph.add_edge(v[0], v[1], weight=v[2])

    return nxdigraph


import random

def is_valid_edge(generated_edges: dict, i: int, j: int):
    return i != j and not generated_edges.get((i, j), None) and not generated_edges.get((j, i), None)

def instance_generator(n: int):
    """
        Input: cantidad de vértices
        Output: una lista que contiene todos los arcos y el número del vértice fuente (la función retorna dos variables).
        Los arcos vienen en la forma (i, j, weight), donde i es el vértice origen del arco y j el vértice al que apunta el arco, mientras que weight es su peso.
    """
    graph = []
    nodes = random.sample(range(0, n), n)
    unvisited_nodes = random.sample(range(0, n), n)
    
    generated_edges = {}
    for i in nodes:
        rand = random.sample(nodes, random.randint(1, 3))

        for j in rand:
            edge = (i, j)
            edge_with_weight = (i, j, random.randint(1, 100))
            
            if generated_edges.get((edge[1], edge[0]), None):
                continue
            
            if i == j:
                new_vertice = None
                iterations = 0
                while new_vertice is None and iterations < 250:
                    iterations += 1
                    number = random.randint(0, n - 1)
                    if is_valid_edge(generated_edges, i, number):
                        new_vertice = number

                if iterations >= 250:
                    return instance_generator(n)
                
                edge = (i, new_vertice)
                edge_with_weight = (i, new_vertice, random.randint(-25, 100)) # -25 y 100 corresponde a los límites de los pesos, puede cambiarlos.
            
            graph.append(edge_with_weight)
            generated_edges[edge] = edge

            if edge_with_weight[1] in unvisited_nodes:
                unvisited_nodes.remove(edge_with_weight[1])

    for i in unvisited_nodes:
        valid_edge = False
        iterations = 0
        while not valid_edge and iterations < 250:
            iterations += 1
            m = random.randint(0, n - 1)
            if is_valid_edge(generated_edges, m, i):
                valid_edge = True
                edge = (m, i)
                edge_with_weight = (m, i, random.randint(-25, 100)) # -25 y 100 corresponde a los límites de los pesos, puede cambiarlos.
                graph.append(edge_with_weight)
                generated_edges[edge] = edge

        if iterations >= 250:
            return instance_generator(n)

    return graph, graph[0][0]

def BellmanFord(grafo, V, E, src,  verbose = False):
	if (verbose == True):
		print("Grafo:", grafo,"Vertices", V, "Nodos", E, "Inicio", src,"\n")
	arrayDistancia = [maxsize] * V
	arrayDistancia[src] = 0
	if (verbose==True): 
		print("Se recorre desde el nodo inicial hacia todos sus adyacentes\n")
  #Se itera V-1 veces
	for i in range(V - 1):
		if (verbose==True):
			print("Nodo: ",i)
		for j in range(E):
			if ((verbose==True)):
				print("Distancia entre",arrayDistancia[grafo[j][0]],"con", grafo[j][2],", si  es menor a ", arrayDistancia[grafo[j][1]])
				if arrayDistancia[grafo[i][0]] + grafo[j][2] < arrayDistancia[grafo[i][1]]:
					print("Es menor, se relajan los nodos\n")
			if arrayDistancia[grafo[j][0]] + grafo[j][2] < arrayDistancia[grafo[j][1]]:
				arrayDistancia[grafo[j][1]] = arrayDistancia[grafo[j][0]] + grafo[j][2]
		if (verbose==True):
			print("\n")	
	for i in range(E):
		x = grafo[i][0]
		y = grafo[i][1]
		peso = grafo[i][2]
		if arrayDistancia[x] != maxsize and arrayDistancia[x] + peso < arrayDistancia[y]:
			if (verbose==True):
				print("El gráfico contiene un ciclo de peso negativo")
	if (verbose==True):
		print("Distancia del vértice al nodo inicial")
		for i in range(V):
			print("%d\t\t%d" % (i, arrayDistancia[i]))
```

# **3. Tiempo de Ejecución**

El tiempo de ejecución del algoritmo dependerá de la cantidad de nodos y la cantidad de arcos del grafo.

En muchos casos se obtendrá el resultado en las primeras iteraciones mientras que en muchos casos se requerirá de hacer todas las iteraciones para llegar al resultado.

**Caso Promedio**: se considera caso promedio cuando el tiempo de ejecución depende de la cantidad de nodos del grafo ($V$) y su cantidad de arcos ($E$) obteniendo así una complejidad temporal $O(E*V)$.

**Peor Caso**: se considera peor caso cuando los nodos del grafo evaluado tienen la mayor cantidad posible de conexiones entre siendo la cantidad de arcor $E$ = (V^2)$.

Sabiendo esto, la complejidad temporal para el peor caso se obtiene como:

$O(E*V) = O(E^2) * O(V) = O(V^3)$

Mejor Caso: se considera mejor caso cuando el grafo va de izquierda a derecha dependiendo así el tiempo de ejecución exclusivamente del número de arcos del grafo. Así se obtiene que la complejidad temporal es $O(E)$.

# **4. Correctitud**

Se probará la correctitud del algoritmo mediante inducción matemática.



**Teorema**

Dado un grafo $G$
$(V.E)$ y un nodo de origen $s$, el algoritmo
Bellman-Ford  encuentra la ruta mas corta a cada nodo $v \in V$ de forma correcta.


**Hipótesis Inductiva**

Despues de cada iteración **i**, para cada nodo $v \in V$,
$D[v,i]$ es el largo de la ruta más corta desde **s** hasta **v** utilizando a lo más $i$ arcos.

**Caso Base**: Desúes de la iteración 0, se tiene que $D[s,0] = 0$ y $D[v,0] = ∞$ para cualquier nodo $v\in V -$ {$s$}.

image.png

**Paso inductivo**: Asumiendo que la hipótesis inductiva se mantiene para
$i$, por lo que se probará que se mantiene para $i+1$

Sea $v$ un **nodo** de $V$, se asume que existe una ruta más corta entre
$s$ y $v$. Sea $u$ el **nodo anterior** a $v$ en esta ruta:

Por hipótesis inductiva, se sabe que $D[u, i]$ es la distancia más corta
entre el **nodo raíz** $s$ y el **nodo** $u$. En la iteración $i+1$,
se cumple que $D[v, i + 1] <= D[u, i] + w(u, v)$ por el paso de reajuste del algoritmo, y
 también que $D[v, i + 1]$ es **mayor o igual** al largo de la ruta mas corta
entre $s$ y $v$ utilizando al menos $i+1$ arcos. Por lo tanto, $D[v,i + 1)$ es el largo de la
ruta mas corta entre $s$ y $v$ utilizando máximo $i+1$ **arcos**.

Después de $n-1 $ iteraciones, para cada nodo $v\in V$, $D[v,n-1]$ es el largo de la ruta más corta desde el nodo raíz $s$ hasta el nodo $v$ utilizando como máximo $n-1$ arcos.
